# Bernoulli Trials with a coin

A Bernoulli trial is an experiment with exactly two possible outcomes. A coin toss is the perfect example for a bernoulli trial. For a fair coin the probabilities of either outcome, _head_ or _tail_ is 0.5.

```python
%pylab inline
plt.rcParams['image.cmap'] = 'gray'
```

Let's model this coin toss with a bit. 1 means heads and 0 means tails. Let's simulate 60 coin tosses. The `np.random.randint()` function samples a discrete uniform distribution. This means that both 0 and 1 are equally likely as for a fair coin toss. Let's visualize this as a one dimensional collection of black and white squares. Each square represents a coin toss.

```python
num_tosses = 50
num_trials = 100
```

```python
tosses = np.random.randint(0,2,size=(1,num_tosses))
fig1,ax1 = plt.subplots(figsize=(20,5))
ax1.imshow(tosses)
ax1.set_xticks(np.arange(0,size(tosses)));
ax1.set_yticks([]);
```

Let's do this 100 times.

```python
trials = np.random.randint(0,2,size=(num_tosses,num_trials))
fig2, ax2 = plt.subplots(figsize=(20,10))
ax2.matshow(trials)
ax2.set_xticks([]);
ax2.set_yticks([]);
ax2.set_xlabel('Trials',fontsize=40);
ax2.set_ylabel('Tosses',fontsize=40);
```

Let's count the number of heads per trial. 

```python
def nCr(n,r):
    return (np.math.factorial(n)/(np.math.factorial(r)*np.math.factorial(n-r)))
```

```python
num_heads = np.count_nonzero(trials,axis=0)
bar_width = (num_heads.max()-num_heads.min())
fig3,(ax3a,ax3b) = plt.subplots(1,2,figsize=(20,5))
n, bins, patches = ax3a.hist(
    num_heads,
    bins=np.arange(-0.5,num_tosses+0.5,1),
    width=1,
    ec='k',
    density=True
)
ax3a.set_xlim(num_heads.min()-1.5, num_heads.max()+1.5);
ax3a.set_xticks(np.arange(num_heads.min(),num_heads.max()+1,step=1))
ax3b.plot(np.arange(num_tosses),n,'o')
ax3b.set_xlim(num_heads.min()-1, num_heads.max()+1);
ax3b.set_xticks(np.arange(num_heads.min(),num_heads.max()+1,step=1));
l = np.arange(num_heads.min(),num_heads.max()+1,step = 1)
binomial_pdf = np.array([nCr(num_tosses,i)/2**num_tosses for i in l])
ax3b.plot(l,binomial_pdf)
```

Now let's tey this with 100 coin tosses per trial for a million trials.
